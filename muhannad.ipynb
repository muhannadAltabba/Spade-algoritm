```python
sequential = [
[["a"], ["a", "b", "c"], ["a", "c"], ["c"]],
[["a"], ["c"], ["b", "c"]],
[["a", "b"], ["d"], ["c"], ["b"], ["c"]],
[["a"], ["c"], ["b"], ["c"]]
]
```

```python
spade = []
```

```python
def addToSpade(item, spade):
    b = False
    for itemes in spade:
        if itemes[0][0][0] == item[0][0]:
            itemes[1].append(item[1])
            itemes[2].append(item[2])
            b = True
    if not b:
        spade.append(([item[0]],[item[1]],[item[2]]))
    return spade
```

```python
def firstScan(sequential):
    arr = []
    for sid in range(len(sequential)):
        for eid in range(len(sequential[sid])):
            for t in sequential[sid][eid]:
                arr = addToSpade([[t], sid + 1, eid + 1], arr)
    return arr
```

result of the first scan

```python
spade = firstScan(sequential)
print(spade)
```

```python
def printPatterns(arr):
    res = []
    for item in arr:
        res.append((item[0], len(set(item[1]))))
    return res
```

```python
#Maintains values >= minsupport
def keepPatterns(arr, minSup):
    t = []
    for item in arr:
        if len(set(item[1])) >= minSup:
            t.append(item)
    return t
```

```python
spade = keepPatterns(spade, 2)
```

```python
def canMarge(l1, l2, commonElements):
    for i in range(commonElements):
        if l1[len(l1) - 1 - i] != l2[commonElements - i - 1 ]:
            return False
    return True
```

```python
def lineToArr(line):
    local = []
    total = []

    for i in range(len(line)):
        if line[i] == ';' and local[0] != '<':
            local = ''.join(local)
            total.append([local])
            local = []
        else:
            local.append(line[i])
    return total
```
